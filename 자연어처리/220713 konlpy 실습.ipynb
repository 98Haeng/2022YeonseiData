```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
text = open('09. alice.txt').read()
```

```python
text
```

```python
wc = WordCloud(background_color='white',max_words=200)
wc=wc.generate(text)
wc.words_
```

```python
# may, year, time, one, House, Case, Power, thereof, every
stopwords=set(STOPWORDS)
stopwords.add('said')
stopwords.add('may')
stopwords.add('year')
stopwords.add('time')
stopwords.add('one')
stopwords.add('House')
stopwords.add('Case')
stopwords.add('Power')
stopwords.add('thereof')
stopwords.add('every')
wc=WordCloud(background_color='white',max_words=2000, stopwords=stopwords)
wc=wc.generate(text)
wc.words_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
alice_mask
```

```python
plt.figure(figsize=(5,5))
plt.imshow(alice_mask, interpolation='bilinear', cmap=plt.cm.gray)
plt.show()
```

```python
wc=WordCloud(background_color='white',max_words=2000, stopwords=stopwords, mask=alice_mask)
wc=wc.generate(text)
```

```python
plt.figure(figsize=(20,15))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# A new hope 스타워즈

text=open('09. a_new_hope.txt').read()
text = text.replace('HAN','Han')
text = text.replace("LUKE 'S'","Luke")
mask=np.array(Image.open('09. stormtrooper_mask.png'))
```

```python
# text 불용어 제거
stopwords = set(STOPWORDS)
```

```python
wc=WordCloud(max_words=1000, stopwords=stopwords, mask=mask,margin=1, random_state=1)
wc=wc.generate(text)
wc
```

```python
wc.words_
```

```python
# vc 이미지 출력, 회색 이미지 출력 옵션 RGB(255,255,255), HSL(A,B, random)
import random
def gray_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return 'hsl(0,0%%,%d%%)'%random.randint(60,100)
    
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```
